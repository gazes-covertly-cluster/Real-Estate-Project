```python
#check Header
df_census2 = pd.read_csv("Census2.csv", header=0)

```

```python
# Keep only the label and Crow Wing County estimate columns
df_crow_wing = df_census2[['Label (Grouping)', 'Crow Wing County, Minnesota!!Total!!Estimate']].copy()

# Rename columns for clarity
df_crow_wing.columns = ['label', 'estimate']
```

```python
df_census2.columns
```

```python
# Strip whitespace and lowercase labels for consistency
df_crow_wing['label'] = df_crow_wing['label'].str.strip().str.lower()

# Convert estimate column to numeric (if it's not already)
df_crow_wing['estimate'] = pd.to_numeric(df_crow_wing['estimate'], errors='coerce')

# Drop rows with missing or zero estimates
df_crow_wing = df_crow_wing[df_crow_wing['estimate'].notnull()]
df_crow_wing = df_crow_wing[df_crow_wing['estimate'] > 0]
```

```python
#Filter the keywords for model
keywords = [
    "total population", "median age", "householder living alone",
    "owner-occupied housing units", "mean earnings", "median value",
    "poverty", "veteran", "disability", "different county"
]

df_crow_wing_filtered = df_crow_wing[df_crow_wing['label'].str.contains('|'.join(keywords))]

```

```python
import pandas as pd

df_scraper = pd.read_csv("RealtorScraper.csv", low_memory=False)
df_scraper.columns[:50]  # See the first 50 column headers
df_scraper.shape  # Check total rows and columns
```

```python
df_scraper.columns = df_scraper.columns.str.strip()  # optional: .str.lower()


columns_to_keep = [
    "address/locality",
    "address/postalCode",
    "address/region",
    "address/street",
    "advertisers/0/broker/name",
    "advertisers/0/email",
    "advertisers/0/name",
    "history/0/event_name",
    "history/0/price",
    "taxHistory/0/assessment/total",
    "BuildingPermitsHistory",
    "BuildingPermitsHistory/0/permit_effective_date",
    "buildingPermitsHistory/0/permit_project_type_1",
    "details/0/category",
    "details/1/category",
]
existing_cols = [col for col in columns_to_keep if col in df_scraper.columns]
df_cleaned = df_scraper[existing_cols].copy()

```
